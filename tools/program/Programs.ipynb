```python
import requests
from requests.auth import HTTPBasicAuth
import os
from dotenv import load_dotenv

# Função para executar o script
def clear_screen():
    os.system('cls')

def run_script():
    clear_screen()
    # Carregue as variáveis de ambiente a partir do arquivo .env
    load_dotenv()

    # Obtenha as variáveis de ambiente
    username = os.getenv("ELOQUA_USERNAME")
    password = os.getenv("ELOQUA_PASSWORD")

    # Configurações da API do Eloqua
    base_url = "https://secure.p04.eloqua.com"
    # Faça a autenticação
    auth = (username, password)

    # Defina a URL para a primeira consulta (usando o ID do programa 2125)
    program_id = "2125"
    url = f"{base_url}/api/REST/2.0/assets/program/{program_id}"

    # Faça a chamada à API
    response = requests.get(url, auth=auth)

    # Verifique se a solicitação foi bem-sucedida
    if response.status_code == 200:
        data = response.json()

        print("'Workshop | Latam Integration 2023'")
        # Procurar o elemento "Error without SFDC ID"
        for element in data.get("elements", []):
            if element.get("name") == "Error without SFDC ID":
                print(f"Dados para:", element.get("name"))
                print("memberCount:", element.get("memberCount"))
                # print("memberErrorCount:", element.get("memberErrorCount"))
                break

        # Procurar o elemento "Error Create Lead"
        for element in data.get("elements", []):
            if element.get("name") == "Error Create Lead":
                print(f"Dados para:", element.get("name"))
                print("memberCount:", element.get("memberCount"))
                # print("memberErrorCount:", element.get("memberErrorCount"))
                break
    else:
        print("Erro na solicitação à API Eloqua. Código de status:", response.status_code)

    print("\n" + "=" * 50 + "\n")

    # Defina a URL para a segunda consulta (usando o ID do programa 1685)
    program_id = "1685"
    url = f"{base_url}/api/REST/2.0/assets/program/{program_id}"
    # Faça a chamada à API
    response = requests.get(url, auth=auth)

    # Verifique se a solicitação foi bem-sucedida
    if response.status_code == 200:
        data = response.json()
        print("'FL_Online_Appt_Routing_2021'")

        # Procurar o elemento "Error without SFDC ID"
        for element in data.get("elements", []):
            if element.get("name") == "Error creating contact":
                print(f"Dados para:", element.get("name"))
                print("memberCount:", element.get("memberCount"))
                # print("memberErrorCount:", element.get("memberErrorCount"))
                break

        # Procurar o elemento "Error Create Lead"
        for element in data.get("elements", []):
            if element.get("name") == "Wait - Account create error":
                print(f"Dados para:", element.get("name"))
                print("memberCount:", element.get("memberCount"))
                # print("memberErrorCount:", element.get("memberErrorCount"))
                break
        
        for element in data.get("elements", []):
            if element.get("name") == "Fix - No Company":
                print(f"Dados para:", element.get("name"))
                print("memberCount:", element.get("memberCount"))
                # print("memberErrorCount:", element.get("memberErrorCount"))
                break
        
        for element in data.get("elements", []):
            if element.get("name") == "Fix - Not US":
                print(f"Dados para:", element.get("name"))
                print("memberCount:", element.get("memberCount"))
                # print("memberErrorCount:", element.get("memberErrorCount"))
                break
    
    else:
        print("Erro na solicitação à API Eloqua. Código de status:", response.status_code)
    print("\n" + "=" * 50 + "\n")

# schedule.every(1).second.do(run_script)

# Executar o script uma vez imediatamente
run_script()

# while True:
#     schedule.run_pending()
#     time.sleep(10)
```
