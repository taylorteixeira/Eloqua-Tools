```python
import os
import pandas as pd
import requests
from requests.auth import HTTPBasicAuth
from tkinter import Tk, filedialog


def ler_email_excel():
    root = Tk()
    root.withdraw()

    file_path = filedialog.askopenfilename()

    df = pd.read_excel(file_path)
    return df  


def obter_autenticacao():
    username = os.getenv("ELOQUA_USERNAME")
    password = os.getenv("ELOQUA_PASSWORD")
    return HTTPBasicAuth(username, password)


def arquivar_emails():
    df = ler_email_excel()

    # Obtém os IDs da coluna "ids"
    ids_column = df['Email ID'].tolist()

    api_url = "https://secure.p04.eloqua.com/api/rest/2.0/assets/email/"

    auth = obter_autenticacao()

    # Realiza um POST para cada ID
    for email_id in ids_column:
        post_url = f"{api_url}{email_id}/archive"
        response = requests.post(post_url, auth=auth)

        if response.status_code == 201:
            print(f"E-mail com ID {email_id} arquivado com sucesso.")
        else:
            print(
                f"Falha ao arquivar e-mail com ID {email_id}. Código de status: {response.status_code}")


if __name__ == "__main__":
  arquivar_emails()
```
