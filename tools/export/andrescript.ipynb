```python
import os
from dotenv import load_dotenv
import requests
from requests.auth import HTTPBasicAuth

load_dotenv()
username = os.getenv("ELOQUA_USERNAME")
password = os.getenv("ELOQUA_PASSWORD")

auth = HTTPBasicAuth(username, password)

def get_template_name(template_id):
    template_url = f"https://secure.p04.eloqua.com/API/REST/2.0/assets/template/{template_id}"
    response = requests.get(template_url, headers=headers, auth=auth)
    template_data = response.json()
    return template_data['name'] if 'name' in template_data else None

 

url = "https://secure.p04.eloqua.com/api/REST/2.0/assets/landingPages?depth=partial&search=currentStatus%3D1"

#url = https://secure.p04.eloqua.com/api/REST/2.0/assets/emails?depth=partial&search=name=Westlaw-Asia-Welcome-2022-EM2

#url = https://secure.p04.eloqua.com/api/REST/2.0/assets/emails?depth=partial&search=createdAt%3E%3D'1672542000'


params = {"page": 1, "pageSize": 1000}

headers = {
    'cookie': "culture=en-US; ELQCOUNTRY=IL; ELOQUA=GUID%3D1A790269C5114C2D9C2EA4CE4475D029; ELQSTATUS=OK",
    # 'Authorization': "Basic dGVtYWlsc1xsYXRhbS5hcGk6RWxvcXVhQDAxMjYxNQ==",
    'content-type': "multipart/form-data; boundary=---011000010111000001101001"
}


while True:
    response = requests.get(url, params=params, headers=headers, auth=auth)
    contentJson = response.json()
    formFields = contentJson['elements']
    print(contentJson['total'])

    
    for data in formFields:
        try:
            template_name = get_template_name(data['sourceTemplateId'])
            print(data['sourceTemplateId'], ",", data['name'], ",", template_name)
        except:
            print('n/a', ",", data['name'], ",", 'n/a')
    
    if contentJson['page'] * contentJson['pageSize'] < contentJson['total']:
        params['page'] += 1
    else:
        break
```
