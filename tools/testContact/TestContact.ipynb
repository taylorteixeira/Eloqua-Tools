```python
import pandas as pd
import requests
from dotenv import load_dotenv
import os
from requests.auth import HTTPBasicAuth
from tkinter import Tk
from tkinter.filedialog import askopenfilename

# Carregar variáveis de ambiente
load_dotenv()

# Obter as variáveis de ambiente
username = os.getenv("ELOQUA_USERNAME")
password = os.getenv("ELOQUA_PASSWORD")

# Configurações da API do Eloqua
base_url = "https://secure.p04.eloqua.com"
contact_endpoint = "/api/REST/1.0/data/contacts"

# Autenticação usando nome de usuário e senha
auth = HTTPBasicAuth(username, password)

# Abrir uma janela de diálogo para selecionar o arquivo Excel
root = Tk()
root.withdraw()  # Oculta a janela principal
file_path = askopenfilename(filetypes=[("Excel files", "*.xlsx")])

# Verificar se um arquivo foi selecionado
if file_path:
    # Usar o pandas para ler o arquivo Excel
    df = pd.read_excel(file_path)

    # Adicionar colunas "Test Migration" e "CS_Existing_Customer" e preenchê-las com "No"
    df["Test Migration"] = "No"
    df["CS_Existing_Customer"] = "No"

    # Iterar sobre os endereços de e-mail na coluna "Email Address"
    for index, row in df.iterrows():
        email = row['Email Address']
        print(f"Consultando API Eloqua para {email}...")

        # Consultar API Eloqua para Test Migration e CS_Existing_Customer
        response = requests.get(
            f"{base_url}{contact_endpoint}?search=name=\"{email}\"&depth=complete",
            auth=auth
        )
        data = response.json()

        # Verificar se o valor é "Test Migration" com o objeto id "100446"
        if any(field.get("id") == "100446" and field.get("value") == "Test Migration"
               for element in data.get("elements", [])
               for field in element.get("fieldValues", [])):
            df.at[index, "Test Migration"] = "Yes"

        # Verificar se o valor é "CS_Existing_Customer" com o objeto id "100495"
        if any(field.get("id") == "100495" and field.get("value") == "CS_Existing_Customer"
               for element in data.get("elements", [])
               for field in element.get("fieldValues", [])):
            df.at[index, "CS_Existing_Customer"] = "Yes"

# Salvar a planilha atualizada
output_path = "output.xlsx"
df.to_excel(output_path, index=False)
print(f"Planilha atualizada salva em {output_path}")
```
